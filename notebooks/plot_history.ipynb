```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
history_path = "../history/inceptionv3_small_11_17_categorical_30.pkl"
with open(history_path, "rb") as f:
    history = pickle.load(f)
```

```python
for k, v in history.items():
    if not k.startswith("val"):
        plt.figure()
        plt.plot(np.arange(1,31), v, label=k)
        plt.plot(np.arange(1,31), history[f"val_{k}"], label=f"val_{k}")
        plt.xlabel("Epochs")
        plt.title(f"Training vs Validation {k.capitalize()}")
        plt.legend()
        plt.savefig(f"../plots/{k.capitalize()}_inceptionv3_small_11_17_categorical_30.png")
```
